```python
import pandas as pd

# Define paths
train_path = "/content/smart_city_traffic/smart-city-traffic-patterns/train_aWnotuB.csv"
test_path = "/content/smart_city_traffic/smart-city-traffic-patterns/datasets_8494_11879_test_BdBKkAj.csv"

# Load datasets
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

# Display first few rows
print("🔹 Training Data:")
display(train_df.head())

print("\n🔹 Testing Data:")
display(test_df.head())
```

```python
print("Train Junctions:", train_df['Junction'].unique())
print("Test Junctions:", test_df['Junction'].unique())
```

```python
from prophet import Prophet
import pandas as pd

# To store final predictions
all_predictions = pd.DataFrame()

# Loop through each junction in TEST DATA
for junction_id in test_df['Junction'].unique():
    print(f"📍 Training for Junction {junction_id}")

    # Filter corresponding training data
    junction_train = train_df[train_df['Junction'] == junction_id][['DateTime', 'Vehicles']]

    if junction_train.empty:
        print(f"⚠️ No training data for Junction {junction_id}. Skipping...")
        continue

    # Rename columns for Prophet
    junction_train.columns = ['ds', 'y']

    # Train the model
    model = Prophet()
    model.fit(junction_train)

    # Filter test data for this junction
    junction_test = test_df[test_df['Junction'] == junction_id].copy()
    junction_test['ds'] = junction_test['DateTime']

    # Forecast future values
    forecast = model.predict(junction_test[['ds']])
    junction_test['Vehicles'] = forecast['yhat'].round().astype(int)

    # Save required output
    all_predictions = pd.concat([all_predictions, junction_test[['ID', 'Vehicles']]])

# Finalize submission
all_predictions = all_predictions.sort_values(by='ID').reset_index(drop=True)
print("✅ Forecasting complete!")

# Preview output
all_predictions.head()
```

```python
import pandas as pd

# Load the training and test datasets
train = pd.read_csv("/content/smart_city_traffic/smart-city-traffic-patterns/train_aWnotuB.csv")
test = pd.read_csv("/content/smart_city_traffic/smart-city-traffic-patterns/datasets_8494_11879_test_BdBKkAj.csv")

# Quick check
print("Train shape:", train.shape)
print("Test shape:", test.shape)
train.head()
```

```python
from prophet import Prophet
import pandas as pd

forecast_df = pd.DataFrame()

for junction_id in train['Junction'].unique():
    print(f"📍 Training Prophet model for Junction {junction_id}")

    # Prepare training data
    train_j = train[train['Junction'] == junction_id].copy()
    train_j['ds'] = pd.to_datetime(train_j['DateTime'])
    train_j['y'] = train_j['Vehicles']
    train_j = train_j[['ds', 'y']]

    # Fit Prophet
    model = Prophet()
    model.fit(train_j)

    # Prepare test data
    test_j = test[test['Junction'] == junction_id].copy()
    test_j['ds'] = pd.to_datetime(test_j['DateTime'])

    # Only keep 'ds' for prediction
    future = test_j[['ds']].copy()
    forecast = model.predict(future)

    # Align predictions with IDs
    test_j['Vehicles'] = forecast['yhat'].iloc[:len(test_j)].round().astype(int)
    forecast_df = pd.concat([forecast_df, test_j[['ID', 'Vehicles']]])

# Sort and reset
forecast_df = forecast_df.sort_values('ID').reset_index(drop=True)

# Check for any NaNs
print("❗ NaN count:", forecast_df['Vehicles'].isna().sum())

# Save file
forecast_df.to_csv("/content/SmartCity_Traffic_Forecast.csv", index=False)
print("✅ Forecast saved.")
```
