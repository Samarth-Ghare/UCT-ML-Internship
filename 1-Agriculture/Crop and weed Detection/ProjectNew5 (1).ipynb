```python
import os
import random
import shutil

# Define paths
img_train_dir = 'data/images/train'
label_train_dir = 'data/labels/train'
img_val_dir = 'data/images/val'
label_val_dir = 'data/labels/val'

# Create val folders if they don't exist
os.makedirs(img_val_dir, exist_ok=True)
os.makedirs(label_val_dir, exist_ok=True)

# List all image files
image_files = [f for f in os.listdir(img_train_dir) if f.endswith('.jpeg')]
val_size = int(0.2 * len(image_files))
val_files = random.sample(image_files, val_size)

# Move files
for img in val_files:
    label = img.replace('.jpeg', '.txt')
    shutil.move(os.path.join(img_train_dir, img), os.path.join(img_val_dir, img))
    shutil.move(os.path.join(label_train_dir, label), os.path.join(label_val_dir, label))

print(f"✅ Moved {val_size} images and labels to validation set.")
```
